```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head(25)
```

```python
# Checking the number of mice.
study_data_complete["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint. Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mouse_ids = study_data_complete.loc[study_data_complete.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mouse_ids

# Code Breakdown:
    # study_data_complete.duplicated(subset=['Mouse ID', 'Timepoint']): This part of the code checks for duplicates in the DataFrame based on the columns 'Mouse ID' and 'Timepoint'. It returns a boolean Series where True indicates a duplicate row and False indicates a unique row.
    # .loc[...]: This is used to access a group of rows and columns by labels or a boolean array. In this case, it filters the DataFrame based on the condition provided inside the square brackets.
    # 'Mouse ID']: This specifies that we are interested in the 'Mouse ID' column from the filtered rows.
    # Variable duplicate_mouse_ids will contain an array of unique mouse IDs that have duplicate entries in the DataFrame based on the 'Mouse ID' and 'Timepoint' columns.
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = study_data_complete.loc[study_data_complete["Mouse ID"]=="g989"]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_study_data_complete = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mouse_ids)==False]
clean_study_data_complete.head(25)

# Code breakdown:
    # study_data_complete["Mouse ID"].isin(duplicate_mouse_ids)==False: This part of the code checks if the 'Mouse ID' in each row of the DataFrame is not present in the duplicate_mouse_ids array.
    # study_data_complete[...]: This filters the study_data_complete DataFrame based on the condition provided inside the square brackets.
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_study_data_complete["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

means = clean_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
medians = clean_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variances = clean_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
sds = clean_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sems = clean_study_data_complete.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_table = pd.DataFrame({"Mean Tumor Volume":means,
                                "Median Tumor Volume":medians,
                                "Tumor Volume Variance":variances,
                                "Tumor Volume Std. Dev.":sds,
                                "Tumor Volume Std. Err.":sems})
summary_table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
summary_table = clean_study_data_complete.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

# This groups the clean_study_data_complete DataFrame by "Mouse ID" and then selects the maximum "Timepoint" value for each group. This effectively finds the last recorded timepoint for each mouse in the study.
max_tumor = clean_study_data_complete.groupby(["Mouse ID"])['Timepoint'].max()

# After obtaining the maximum timepoints for each mouse, this line resets the index of the resulting DataFrame max_tumor. By resetting the index, the "Mouse ID" and the corresponding maximum "Timepoint" values are converted into columns in the DataFrame, and a new default integer index is assigned.
max_tumor = max_tumor.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = max_tumor.merge(clean_study_data_complete,on=['Mouse ID','Timepoint'],how="left")

df_blah = pd.DataFrame(merged_data)
df_blah
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatment_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_list.append(final_tumor_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    # Find outliers and print
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
orange_out = dict(markerfacecolor='red',markersize=12)
plt.boxplot(tumor_vol_list, labels = treatment_list,flierprops=orange_out)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

capomulin_table = clean_study_data_complete.loc[clean_study_data_complete['Drug Regimen'] == "Capomulin"]
mousedata = capomulin_table.loc[capomulin_table['Mouse ID']== 'l509']

plt.plot(mousedata['Timepoint'],mousedata['Tumor Volume (mm3)'])
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Assuming you have the necessary data loaded and processed

# Calculate the average tumor volume per mouse for the Capomulin regimen
capomulin_avg_tumor_volume = capomulin_table.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
capomulin_avg_tumor_volume = capomulin_avg_tumor_volume.reset_index()

# Merge the average tumor volume data with the mouse weight data
capomulin_avg_tumor_volume = pd.merge(capomulin_avg_tumor_volume, mouse_metadata, on='Mouse ID')

# Create a scatter plot
plt.figure(figsize=(10,6))
plt.scatter(capomulin_avg_tumor_volume['Weight (g)'], capomulin_avg_tumor_volume['Tumor Volume (mm3)'], color='blue', alpha=0.7)
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
# Scatter plot
# Extract mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import pearsonr
mouse_weight = capomulin_avg_tumor_volume['Weight (g)']
avg_tumor_volume = capomulin_avg_tumor_volume['Tumor Volume (mm3)']

# Calculate the correlation coefficient
correlation_coefficient, _ = pearsonr(mouse_weight, avg_tumor_volume)
print(f"Correlation Coefficient: {correlation_coefficient}")

from scipy.stats import linregress

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(mouse_weight, avg_tumor_volume)

# Create the equation of the line
line_eq = f"y = {slope:.2f}x + {intercept:.2f}"
print(f"Linear Regression Model: {line_eq}")
print(f"R-squared Value: {r_value**2}")

plt.figure(figsize=(10,6))
plt.scatter(mouse_weight, avg_tumor_volume, color='blue', alpha=0.7)

# Regression line
regress_values = mouse_weight * slope + intercept
plt.plot(mouse_weight, regress_values, color='red', linestyle='--')

# Annotate the line equation
plt.annotate(line_eq, (20, 40), fontsize=12, color='red')

plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```
